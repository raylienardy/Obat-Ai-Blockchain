```python
# ============================================================
# Phase 2A - Evaluasi Baseline dengan Ground Truth CSV
# ============================================================

base = '/content/drive/MyDrive/BlockchainAI'
gt_path = os.path.join(base, 'demo_test_ground_truth.csv')
demo_dir = os.path.join(base, 'demo_test')
model_path = os.path.join(base, 'models', 'detector_model.pt')

gt = pd.read_csv(gt_path)
gt_clean = gt[gt['category'].isin(['obat', 'non_obat'])]

obat_files = gt_clean[gt_clean['category'] == 'obat']['filename'].tolist()
non_obat_files = gt_clean[gt_clean['category'] == 'non_obat']['filename'].tolist()

print(f"Jumlah gambar obat: {len(obat_files)}")
print(f"Jumlah gambar non-obat: {len(non_obat_files)}")

model = YOLO(model_path)
print("Model detektor dimuat.")

# Evaluasi obat
detected_obat = 0
failed_obat = []
for fname in obat_files:
    path = os.path.join(demo_dir, fname)
    results = model(path, conf=0.5, verbose=False)
    if len(results[0].boxes) > 0:
        detected_obat += 1
    else:
        failed_obat.append(fname)

# Evaluasi non-obat
false_pos = 0
false_pos_files = []
for fname in non_obat_files:
    path = os.path.join(demo_dir, fname)
    results = model(path, conf=0.5, verbose=False)
    if len(results[0].boxes) > 0:
        false_pos += 1
        false_pos_files.append(fname)

recall = detected_obat / len(obat_files) if len(obat_files) > 0 else 0
fp_rate = false_pos / len(non_obat_files) if len(non_obat_files) > 0 else 0

print("\n========== BASELINE EVALUATION (Phase 1b Model) ==========")
print(f"Recall (obat terdeteksi): {recall:.2%} ({detected_obat}/{len(obat_files)})")
print(f"False Positive (non‑obat terdeteksi): {false_pos} dari {len(non_obat_files)} ({fp_rate:.2%})")
print(f"\nObat yang tidak terdeteksi ({len(failed_obat)}):")
for f in failed_obat:
    print(f"  - {f}")
print(f"\nNon‑obat yang salah terdeteksi ({len(false_pos_files)}):")
for f in false_pos_files:
    print(f"  - {f}")
print("============================================================")
```
